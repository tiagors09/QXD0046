1. Construa um AFD que reconhece a linguagem $\{ w \in \{a,b\}^* ~|~ \text{ toda posição ímpar de } w \text{ é } b\}$.

```python
# Insira aqui o código do seu AFD

Q = {1, 2, 3}
Sigma = {'a', 'b'}
delta = {
    (1, 'a'): 3,
    (1, 'b'): 2,
    (2, 'a'): 1,
    (2, 'b'): 3
}
q0 = 1
F = {1, 2}

M = AFD(Q,Sigma,delta,q0,F)
M
```

```python
casos = [
    "abbab",
    "aabbaab",
    "b",
    "ababab",
    "baba",
    "aaaabbbb",
    "bab",
    "aaabb",
    "aabb",
    "baabba"
]

for caso in casos:
  print(f"{caso} é aceito? {M.aceita(caso)}")
```

4. Construa um AFD que reconhece a linguagem $\{ w \in \{a,b\}^* ~|~ w \text{ contém pelo menos dois } a \text{ e no máximo três } b\}$.

6. Construa um AFN que reconhece a linguagem $\{ w \in \{a,b\}^* ~|~ w \text{ contém um número par de } a \text{ ou contém exatamente dois } b\}$ com seis estados.

```python
# Insira aqui o código do seu AFN

Q = {1,2,3,4,5,6,7}
Sigma = {'a','b'}
delta = {
    (1,''):{2,3},
    (2,''):{6},
    (2,'a'):{4},
    (3,'a'):{3},
    (3,'b'):{5},
    (4,'a'):{6},
    (5,'a'):{5},
    (5,'b'):{7},
    (6,''):{2},
    (7,'a'):{7}
}
q0 = 1
F = {6,7}

M = AFN(Q,Sigma,delta,q0,F)
M
```

```python
test_cases = [
    "",         # Não atende a regra (nenhum a, nenhum b)
    "a",        # Não atende a regra (1 a, nenhum b)
    "b",        # Não atende a regra (nenhum a, 1 b)
    "aa",        # Atende a regra (2 a's, nenhum b)
    "ab",        # Atende a regra (1 a, 1 b)
    "ba",        # Atende a regra (1 a, 1 b)
    "bb",       # Não atende a regra (nenhum a, 2 b's)
    "aaa",       # Atende a regra (3 a's, nenhum b)
    "aab",       # Atende a regra (2 a's, 1 b)
    "aba",       # Atende a regra (2 a's, 1 b)
    "abb",      # Não atende a regra (1 a, 2 b's)
    "baa",       # Atende a regra (2 a's, 1 b)
    "bab",      # Não atende a regra (1 a, 2 b's)
    "bba",      # Não atende a regra (1 a, 2 b's)
    "bbb",      # Não atende a regra (nenhum a, 3 b's)
    "aaaa",      # Atende a regra (4 a's, nenhum b)
    "aabb",      # Atende a regra (2 a's, 2 b's)
    "abab",      # Atende a regra (2 a's, 2 b's)
    "abba",     # Não atende a regra (2 a's, 2 b's)
    "abbb",      # Não atende a regra (1 a, 3 b's)
]

# Imprime os resultados
for string in test_cases:
    print(f"{string} => {M.aceita(string)}")
```

7. Construa um AP que reconhece a linguagem $\{a^nb^{n+m}a^m ~|~ m,n \geq 0\}$

```python
# Insira aqui o código do seu AP

Q = {'A','B','C','D'}
Sigma = {'a','b'}
Gamma = {'x','$'}
delta = {
    ('A','',''):{('B','$')},
    ('B','a','$'):{('B','x')},
    ('B','a','x'):{('B','x')},
    ('B','b','x'):{('C','')},
    ('C','a','x'):{('C','x')},
    ('C','b','x'):{('C','')},
    ('C','','$'):{('D','')},
}
q0 = 'A'
F = {'D'}

M = AP(Q,Sigma,Gamma,delta,q0,F)
M
```
