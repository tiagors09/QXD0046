# Exercícios sobre Autômatos Finitos

1. Construa um AFD que reconhece a linguagem $\{ w \in \{a,b\}^* ~|~ \text{ toda posição ímpar de } w \text{ é } b\}$.

```python
# Insira aqui o código do seu AFD

Q = {1, 2}
Sigma = {'a', 'b'}
delta = {
    (1, 'b'): 2,
    (2, 'a'): 1
}
q0 = 1
F = {1, 2}

M = AFD(Q,Sigma,delta,q0,F)
M
```

```python
casos = [
    "abbab",
    "aabbaab",
    "b",
    "ababab",
    "baba",
    "aaaabbbb",
    "bab",
    "aaabb",
    "aabb",
    "baabba"
]

for caso in casos:
  print(f"{caso} é aceito? {M.aceita(caso)}")
```
